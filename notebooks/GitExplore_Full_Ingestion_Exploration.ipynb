```python
from git import Repo
from pathlib import Path

def should_include(file: Path):
  EXTENSIONS = [".py", ".java", ".js", ".md", ".txt", ".c", ".cpp", ".cs"]


  IGNORE_DIRS = {
      ".git",
      "__pycache__",
      "node_modules",
      "venv",
      "dist",
      "build",
      ".github",
      "tests",
  }

  IGNORE_FILES = {
      "package-lock.json",
      "yarn.lock",
  }


  if not file.is_file():
      return False

  if file.suffix not in EXTENSIONS:
      return False

  if any(part in IGNORE_DIRS for part in file.parts):
      return False

  if file.name in IGNORE_FILES:
      return False

  if "test" in file.name.lower():
      return False

  if file.stat().st_size < 0:
      return False

  if file.stat().st_size > 500_000:  # avoid huge files
      return False

  return True

def classify_file(file):
    match file.suffix:

      case ".md":
        return "documentation"

      case ".py":
        return "python code"

      case ".java":
        return "java code"

      case ".js":
        return "javascript code"

      case ".cpp":
        return "c++ code"

      case ".c":
        return "c code"

      case ".cs":
        return "c# code"

      case _:
        return "other"

def generate_content_metadata(file_path: Path, repo_path: Path, repo_url: str, file_classification: dict):
    relative_path = file_path.relative_to(repo_path)
    stat = file_path.stat()

    try:
        content = file_path.read_text(encoding="utf-8")
    except:
        content = None

    return {
        "file_path": str(relative_path),
        "content": content,
        "type_of_file": file_classification[file_path],
        "size_of_file": stat.st_size,
        "last_modified": stat.st_mtime,
        "repo_url": repo_url
    }


def full_github_repo_access(repo_url: str, save_path: str):

    Repo.clone_from(url=repo_url, to_path=save_path)

    repo_path = Path(save_path)

    files = [f for f in repo_path.rglob("*") if should_include(f)]

    file_classification = {}
    for file_path in files:
        file_classification[file_path] = classify_file(file_path)

    for file_path in files:
        metadata = generate_content_metadata(file_path, repo_path, repo_url, file_classification)
        print("\n\n")

        for key, value in metadata.items():
          print(f"{key} - {value}")
```

```python
repo_url = "https://github.com/malivinayak/Multi-Programming"
save_path = "/content/repo2"

full_github_repo_access(repo_url, save_path)

!rm -rf "/content/repo2"
```
